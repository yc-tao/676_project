# (C) Dataset Selection

### Why SatHealth
Primary reason (fit to my interests):
- My research focus is machine learning for healthcare, and SatHealth directly targets public health and disease risk modeling using real-world environmental signals.

Favorable dataset characteristics:
- Clear, high-impact targets: supports predicting social determinants of health (SDI) and disease prevalence/risk, which are meaningful and interpretable healthcare outcomes.
- Multimodal + spatiotemporal structure: combines satellite imagery with environmental variables (e.g., climate/weather) and health outcomes, enabling both classic data mining (associations across regions/time) and modern ML (multimodal representation learning).
- Multiple granularities of analysis: enables regional-level analyses (population health) and individual-level risk modeling (personal health), useful for comparing modeling assumptions and evaluation setups.
- Natural avenues for methodological comparison: easy to compare
  - unimodal vs multimodal models (imagery-only vs environment-only vs fused),
  - static vs temporal models (single snapshot vs sequences),
  - predictive modeling vs descriptive mining (prediction vs interpretable drivers).
- Practical relevance and coherent narrative: environmental exposure → health outcomes is a defensible story for a report and supports insights beyond benchmark scores.

### What I can mine / analyze with it
- Clustering / regional stratification: group regions by environmental signatures and relate clusters to SDI / prevalence.
- Anomaly detection: identify regions/time periods with unusual divergence between environmental signals and observed prevalence.
- Spatiotemporal prediction: forecast prevalence/SDI trends using lagged environmental variables.
- Representation learning: learn embeddings from imagery + environment, then test whether embeddings improve downstream health prediction.

### Trade-offs (vs other candidate datasets)
Compared to UQABench (e-commerce click logs):
- Pros of SatHealth: directly healthcare-relevant; outcomes are societally meaningful; easier to justify in a healthcare ML report.
- Cons of SatHealth: weaker support for classic course topics like frequent itemsets and association rules; less naturally aligned with sequential pattern mining.
- What I lose by not choosing UQABench: a clean setup for sequential behavior modeling (next-action prediction, sequential patterns) and extremely large-scale interaction logs.

Compared to fairness-focused graph benchmarks (credit/recidivism/social graphs):
- Pros of SatHealth: centered on health outcomes rather than adjudication labels (credit/bail); multimodal rather than purely relational; can study environment-driven health disparities.
- Cons of SatHealth: fairness analysis is less “off-the-shelf” because protected groups may not be explicitly represented; requires careful definition of groups and fairness metrics.
- What I lose by not choosing graph benchmarks: standardized fairness tasks, sensitive attributes, and established fairness baselines/metrics for direct comparison.

# (E) Initial Insights and Direction

Observation: Environmental tables (airquality, climate, greenery) are county-level panels with 7,392 rows, consistent with 88 counties × 7 years × 12 months.
Hypothesis: Temporal dynamics and seasonality will be strong drivers; models that ignore time (static averages) will underperform and may misattribute effects.

Observation: Missingness is concentrated in a few variables: NDVI is missing ~49%, NDVI_binary ~32%, and NO2 ~36%, while most climate variables appear complete.
Hypothesis: Naive row-wise deletion will disproportionately remove greener/remote or cloud-prone observations and bias downstream associations; robust imputation or modeling missingness explicitly will be necessary.

Observation: Several airquality variables are uninformative in the current extract (PM2.5 is all zeros), and some variables have very limited variance (NO2 values clustered tightly).
Hypothesis: Some features may be preprocessing artifacts or unit/scale issues; feature validation and variance-based filtering will materially improve model stability.

Observation: Climate features are highly correlated (e.g., latent heat flux vs total evaporation ~1.0; temperature vs soil temperature ~0.996; radiation sums tightly correlated).
Hypothesis: Multicollinearity will inflate coefficients and reduce interpretability in linear models; dimensionality reduction (PCA) or regularization (ridge/lasso) or targeted feature selection will be needed for interpretable inference.

Observation: Landcover and SDI are cross-sectional at the county level (88 rows), while disease prevalence is at CBSA level (47 units) with time (1,969 rows).
Hypothesis: Mismatched geographic scales (CBSA vs county) will be a key bottleneck; analysis must either harmonize to a common geography or use hierarchical/multi-level approaches to avoid ecological fallacy.

Observation: Disease prevalence shows large differences across ICD-10 level-1 codes (e.g., Z00-Z99 and R00-R99 much higher than many others).
Hypothesis: Broad ICD group prevalence may reflect healthcare utilization and coding practices as much as underlying disease burden; focusing on a subset of clinically grounded categories (e.g., cardio/metabolic/respiratory) may yield cleaner signals.

Observation: Google Map points are very large (432,918 points) and link to 88 counties; this is the main high-volume modality compared to the tabular panels.
Hypothesis: Imagery-derived embeddings may add value beyond climate/airquality, but computational cost will be dominated by image feature extraction; sampling strategies or precomputed embeddings will be important for feasibility.

Potential RQs:
- How does handling missingness (imputation vs missing-indicator vs model-based) change estimated relationships between greenery (NDVI) and disease prevalence/SDI?
- Do temporal models (lagged climate/airquality, seasonal terms) outperform static annual averages for predicting prevalence and SDI?
- Which feature sets contribute most: climate-only vs airquality-only vs greenery-only vs landcover+SDI vs multimodal fusion including imagery embeddings?
- How robust are associations to multicollinearity controls (regularization vs PCA vs manual feature pruning based on correlation)?
- What is the impact of geographic harmonization choice (county aggregation to CBSA vs CBSA disaggregation assumptions) on conclusions about environment–health links?
- Are there detectable “outlier” counties/years where prevalence deviates from what environmental variables would predict, suggesting unmeasured drivers (access, policy, coding)?
