|-------|-------------|

| **Target variable(s)** | **Binary node classification** labels per dataset (task-specific; varies by dataset). |
| **Licensing or usage constraints** | Uses multiple **public datasets plus two constructed ones (AMiner-S/L)**; follow each underlying dataset’s original license/terms as stated by their sources. |



### Dataset: UQABench: Evaluating User Embedding for Prompting LLMs in Personalized Question Answering
https://arxiv.org/abs/2502.19178

| Field | Description |
|-------|-------------|
| **Dataset name and source** | Liu, Langming, et al. *“UQABench: Evaluating User Embedding for Prompting LLMs in Personalized Question Answering.”* https://arxiv.org/pdf/2502.19178 |
| **Course topic alignment** | Frequent user–item interaction / sequence data (action prediction), text mining (QA-form evaluation) |
| **Potential beyond-course techniques** | User embedding learning (sequential recommenders/encoders); soft-prompting LLMs with embeddings; aligning embedding space to LLM semantic space |
| **Dataset size and structure** | Source click logs: **31,317,087 user–item interactions**, split **9:1** train/test (**28,094,799 / 3,222,288**). From the test set they generate **7,192 personalized Q&A** for evaluation. |
| **Data types** | User interaction sequences (click logs) + item catalog/attributes; plus generated **natural-language questions** and **answers** for evaluation. |
| **Target variable(s)** | Task-dependent: **sequence understanding** (questions about interaction history); **action prediction** (**next-item** / **next-attribute** as QA); **interest perception** (infer long-/short-term interests and their evolution). |
| **Licensing or usage constraints** | Apache license 2.0 |
